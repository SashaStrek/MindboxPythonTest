```python
#Задание:
#Написать функцию, которая подсчитывает число покупателей, попадающих в каждую группу, 
#если нумерация ID сквозная и начинается с 0.
#На вход функция получает целое число n_customers (количество клиентов).

import pandas

IdList0 = [0, 1, 11, 11, 111, 1111, 11111, 111111, 1111111, 11111111, 111111111, 111111111, 91, 19, 99] #список с ID

print('введите количество пользователей n_customers')
n_customers = int(input()) #ввод количества пользователей

def NumIdInGroup(n_customers):
    WorkList = IdList0[:n_customers]   # указываем, с чем работаем
       #считаем номер группы для id
    GroupList = []
    GroupListUV = []
    NumIdInGroupList = []
    for val in WorkList:
        sum = 0
        while val > 0:
            digit = val % 10
            sum += digit
            val = val // 10
        GroupList.append(sum)
        if sum not in GroupListUV:   # Находим уникальные значения групп
            GroupListUV.append(sum)
            NumIdInGroupList.append(1) # +1 элемент, если новый элемент
        else:
            NumIdInGroupList[GroupListUV.index(sum)] += 1    # +1 к соответствующему количеству, если элемент уже есть
        
    df1 = pandas.DataFrame({'Номер группы': GroupListUV,
                           'Число покупателей': NumIdInGroupList})
    return df1

NumIdInGroup(n_customers)
```
